```python
import time
import random
import matplotlib.pyplot as plt
import heapq

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

def heap_sort(arr):
    heapq.heapify(arr)
    sorted_arr = []
    while arr:
        sorted_arr.append(heapq.heappop(arr))
    return sorted_arr

def measure_time(sort_func, n):
    arr = random.sample(range(n*10), n)
    start_time = time.time()
    sort_func(arr)
    end_time = time.time()
    return end_time - start_time

n_values = [10, 100, 1000, 5000, 10000, 50000, 100000]
times_insertion = []
times_heap = []

for n in n_values:
    times_insertion.append(measure_time(insertion_sort, n))
    times_heap.append(measure_time(heap_sort, n))

plt.plot(n_values, times_insertion, label="Insertion Sort")
plt.plot(n_values, times_heap, label="Heap Sort")
plt.xlabel('Number of elements (n)')
plt.ylabel('Time (seconds)')
plt.title('Time taken for Sorting (Insertion Sort vs Heap Sort)')
plt.legend()
plt.show()
```
